```python
import os
import time
import zipfile
```

```python
file_loc = "/content/5m-Sales-Records.zip"
```

```python
with zipfile.ZipFile(file_loc, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
file_size = os.path.getsize(file_loc) >> 20
print ("Size of file is {} MB".format(file_size))
```

# Pandas

```python
with pd.option_context("mode.dtype_backend", 
                       "pandas"):
    
    start_time = time.time()
    
    df = pd.read_csv("/content/5m Sales Records.csv", 
                     use_nullable_dtypes=True)
    
    end_time = time.time()
    
    result = round(end_time - start_time, 2)
    
    print("Pandas Loading Time = {}".format(result))
```

# Pyarrow

```python
with pd.option_context("mode.dtype_backend", 
                       "pyarrow"):
    
    start_time = time.time()
    
    df = pd.read_csv("/content/5m Sales Records.csv", 
                     use_nullable_dtypes=True, 
                     engine="pyarrow")
    
    end_time = time.time()
    
    result = round(end_time - start_time, 2)
    
    print("Pyarrow Loading Time = {}".format(result))
```
